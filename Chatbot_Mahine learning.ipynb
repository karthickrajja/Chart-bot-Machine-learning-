# Step_2 >> Processing the raw data

```python
lem = nltk.WordNetLemmatizer()

def lemizer(tokens):
    return [lem.lemmatize(token) for token in tokens]

# Removing punctuations
pun_remover = dict((ord(punct),None) for punct in string.punctuation)
```

```python
string.punctuation
```

```python
# Lemnormalize and tokenizing and removing the puncation
def lemnormalize(text):
    return lemizer(nltk.word_tokenize(text.lower().translate(pun_remover)))
```

## Step_3 >> Greating response

```python
Inputs = ("hi","What's up", "hello","hey")
Responses = ("Hi","Hello","I'm excited to talk with you", "How can I help you")
consurn_input = ("How are you")
consurn_output = ("I'm good, how about you")

def greeting(sentence):
    for i in sentence.split():
        if i.lower() in Inputs:
            return random.choice(Responses)
        elif i.lower() in consurn_input:
            return(consurn_output)
```

```python
greeting("Hi hello")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
### Writing Response Function

def response(user_response):
    robo_response=" "
    token_sen.append(user_response)
    tf1 = TfidfVectorizer(tokenizer=lemnormalize, stop_words="english")
    tf1fit = tf1.fit_transform(token_sen)
    vals = cosine_similarity(tf1fit[-1],tf1fit)  
    idx = vals.argsort()[0][-2]
    flat = vals.flatten()
    flat.sort()
    req_tf1fit = flat[-2]
    if (req_tf1fit==0):
        robo_response = robo_response+ "I sorry, I couldn't understand"
        return robo_response
    else:
        robo_response = robo_response+ token_sen[idx]
        return robo_response
    
```

# Start and End Point

```python
flag = True
print("Chatbot: U can chat with me about Machine learning")
while(flag==True):
    user_response = input()
    user_response = user_response.lower()
    if (user_response!="bye"):
        if (user_response =="thanks" or user_response== "thank you" ):
            print("SampleBot: You're welcome")
            flag==False
        else:
            if(greeting(user_response)!=None):
                print("SampleBot:"+greeting(user_response))
            else:
                print("SampleBot" , end = "")
                print(response(user_response))
                token_sen.remove(user_response)
    else:
        flag=False
        print("Thanks, It was pleasant chating with u")
```
